Проверим все числовые признаки на нормальность распределения. Для этого проверим каждый из них 3 тестами: Shapiro–Wilk, D’Agostino, Anderson–Darling на уровне значимости α=0.05.

```python
# Словарь датафреймов
dfs = {
    'market_file': market_file,
    'market_money': market_money,   
    'market_time': market_time,     
    'money': money                  
}

results = []

for name, df in dfs.items():
    # Выбираем все числовые признаки, кроме id
    num_cols = df.select_dtypes(include=['number']).columns.drop('id', errors='ignore')
    for col in num_cols:
        data = df[col].dropna()
        # 1) Shapiro–Wilk (для n ≤ 5000)
        stat_sw, p_sw = shapiro(data)
        # 2) D’Agostino’s K² (нужен минимум ~20 наблюдений)
        stat_dk, p_dk = normaltest(data)
        # 3) Anderson–Darling
        ad = anderson(data, dist='norm')
        # находим критическое значение для α=5%
        try:
            crit_5 = ad.critical_values[list(ad.significance_level).index(5.0)]
        except ValueError:
            # если точного 5% нет, берём ближайшее
            crit_5 = ad.critical_values[2]
        
        results.append({
            'dataset': name,
            'feature': col,
            'shapiro_p': p_sw,
            'dagostino_p': p_dk,
            'anderson_stat': ad.statistic,
            'anderson_crit_5%': crit_5,
            # считаем нормальным, если p > 0.05 или stat < crit
            'normal_shapiro':  p_sw  > 0.05,
            'normal_dagostino': p_dk > 0.05,
            'normal_anderson':  ad.statistic < crit_5
        })

# Собираем результаты в DataFrame
norm_df = pd.DataFrame(results)

# Считаем, сколько признаков в каждом датасете прошло все три теста
summary = (
    norm_df
    .assign(all_normal = lambda d: d[['normal_shapiro','normal_dagostino','normal_anderson']].all(axis=1))
    .groupby('dataset')
    .agg(
        n_features        = ('feature','nunique'),
        n_normal_all_tests= ('all_normal','sum')
    )
)

print("Результаты формальной проверки нормальности (α=0.05):")
display(summary)
```

Лишь один признак прошел все три теста. Это говорит о том, что большинство признаков имеют выраженную асимметрию, выбросы или мультимодальность, поэтому прямое применение методов, требующих нормальности (например, классические t-тесты или linear models) может быть не оптимально.
Для непрерывных, но ненормальных признаков можно рассмотреть лог- трансформацию.

### Использование пайплайнов

**Примените все изученные модели. Для этого используйте пайплайны.**

#### Во время подготовки данных используйте ColumnTransformer. Количественные и категориальные признаки обработайте в пайплайне раздельно. Для кодирования категориальных признаков используйте как минимум два кодировщика, для масштабирования количественных — как минимум два скейлера. Напоминаем, что для каждой модели можно подготовить данные с разным кодированием и масштабированием.

```python
# Разбиентие датафрейма на обучающую и тестовую выборки с удалением столбца с целевым признаком и неинформативного столбца "id"
X_train, X_test, y_train, y_test = train_test_split(
    df_all.drop(['активность','id'], axis=1),
    df_all['активность'],
    test_size = TEST_SIZE, 
    random_state = RANDOM_STATE,
    stratify = df_all['активность'])

# Проверка разбиения на обучающую и тестовую выборки
X_train.shape, X_test.shape
```

В обучающей выборке 972 строк и 16 столбцов, в тестовой - 324 строки и 16 столбцов. Разбиение прошло корректно.

```python
# создаём списки с названиями признаков
ohe_columns = ['информирование', 'популярная_категория']
ord_columns = ['сервис']
num_columns = ['актив_6_мес', 'актив_тек_мес', 'длительность', 'акционные_покупки', 'категорий_за_визит', \
               'неоплаченные_продукты', 'ошибка', 'страниц_за_визит', 'предыдущий_месяц_x', 'препредыдущий_месяц', \
               'текущий_месяц_x', 'предыдущий_месяц_y', 'текущий_месяц_y']
```

```python
# создаём пайплайн для подготовки признаков из списка ohe_columns: заполнение пропусков и OHE-кодирование
# SimpleImputer + OHE
ohe_pipe = Pipeline(
    [('simpleImputer_ohe', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
     ('ohe', OneHotEncoder(handle_unknown="ignore", sparse_output=False))
    ]
    )
```

```python
# создаём пайплайн для подготовки признаков из списка ord_columns: заполнение пропусков и OE-кодирование
# OrdinalEncoder + SimpleImputer
ord_pipe = Pipeline(
    [('simpleImputer_before_ord', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
     ('ord',  OrdinalEncoder(
                categories=[
                    ['стандарт', 'премиум'], 
                ], 
                handle_unknown='use_encoded_value', unknown_value=np.nan
            )
        ),
     ('simpleImputer_after_ord', SimpleImputer(missing_values=np.nan, strategy='most_frequent'))
    ]
)
```

```python
data_preprocessor = ColumnTransformer(
    [
        ('ohe',    ohe_pipe,    ohe_columns),
        ('ord',    ord_pipe,    ord_columns),
        ('std',    StandardScaler(), num_columns),
        ('minmax', MinMaxScaler(),   num_columns),
    ],
    remainder='drop'  # все остальные столбцы (id, активность) отпадут
)
```

```python
# создаём итоговый пайплайн: подготовка данных и модель
pipe_final = Pipeline([
    ('preprocessor', data_preprocessor),
    ('model',        DecisionTreeClassifier(random_state=RANDOM_STATE))
])
```

```python
# Создаем словари для 4 моделей
param_grid = [
    
    # 1) k-Nearest Neighbors
    {
        'model': [KNeighborsClassifier()],
        'model__n_neighbors': [3, 7, 11], # (число соседей)
        'model__weights': ['uniform', 'distance'], # равные веса и взвешивание по расстоянию
        'preprocessor__std':    [StandardScaler(), 'passthrough'],
        'preprocessor__minmax': [MinMaxScaler(),   'passthrough'],
    },

    # 2) Logistic Regression
    {
        'model': [
            LogisticRegression(
                random_state=RANDOM_STATE,
                solver='liblinear',
                penalty='l1',
                max_iter=1000
            )
        ],
        'model__C': range(1, 6),                   # сила регуляризации от 1 до 5
        'preprocessor__std':    [StandardScaler(), 'passthrough'],
        'preprocessor__minmax': [MinMaxScaler(),   'passthrough'],
    },

    # 3) Support Vector Classifier
    {
        'model': [
            SVC(
                random_state=RANDOM_STATE,
                probability=True
            )
        ],
        'model__C': range(1, 6),                   # сила регуляризации от 1 до 5
        'model__kernel': ['linear', 'rbf'],        # линейное и RBF-ядро
        'model__gamma': ['scale', 'auto'],         # два стандартных способа расчёта gamma
        'preprocessor__std':    [StandardScaler(), 'passthrough'],
        'preprocessor__minmax': [MinMaxScaler(),   'passthrough'],
    },
    
    # 4) Decision Tree
    {
        'model': [DecisionTreeClassifier(random_state=RANDOM_STATE)],
        'model__max_depth': range(2, 6),           # глубина деревьев от 2 до 5
        'model__max_features': range(2, 6),        # при разбиении рассматриваем от 2 до 5 признаков
        'model__criterion': ['gini', 'entropy'],  # два основных критерия качества разбиения
        'preprocessor__std':    [StandardScaler(), 'passthrough'],
        'preprocessor__minmax': [MinMaxScaler(),   'passthrough'],
    }
]
```

Объяснение выбора гиперпараметров

DecisionTreeClassifier

max_depth=2…5: даёт контроль над сложностью дерева (мелкие деревья проще и устойчивее).

max_features=2…5: ограничивает число признаков при каждом разбиении, что может предотвратить переобучение.

criterion=['gini','entropy']: проверяем оба классических критерия качества разбиения.


KNeighborsClassifier

n_neighbors=2,4,6,8,10: пробуем от совсем «близкого» соседа до небольшого окружения.

weights=['uniform','distance']: выясним, выгодно ли взвешивать соседей по близости.


LogisticRegression

C=1…5: простая равномерная сетка для силы регуляризации (большие C слабее регуляризуют, малые C сильнее).

Остаются l1 и liblinear, подходящие для небольшого числа признаков.


SVC

C=1…5: аналогично логрегрессии.

kernel=['linear','rbf']: проверяем оба распространённых ядра.

gamma=['scale','auto']: два стандартных способа расчёта параметра RBF-ядра.

preprocessor__num во всех словарях
— пробуем три стратегии для числовых фич: StandardScaler(), MinMaxScaler() и вообще без масштабирования ('passthrough'), чтобы понять, насколько важен скейлинг для каждой модели.

```python
# Создаём объект RandomizedSearchCV для автоматического подбора оптимальных гиперпараметров для всего конвейера pipe_final
randomized_search = RandomizedSearchCV(
    pipe_final, 
    param_grid,
    n_iter=10,     # сокращаем до 10 итераций вместо полного перебора для экономии времени
    cv=3,          # сокращаем число фолдов до 3 вместо 5 для экономии времени
    scoring='roc_auc',
    random_state=RANDOM_STATE,
    # verbose=2,     # в консоль будут выводиться сообщения о каждом пробном сочетании
    n_jobs=-1
    
)
randomized_search.fit(X_train, y_train)
```

```python
# Берём лучшую модель, найденную на CV среди всех четырёх
best_model = randomized_search.best_estimator_

print("Лучшая модель по CV и её параметры:\n", best_model)
print("ROC-AUC на кросс-валидации:", randomized_search.best_score_)

# Оцениваем именно её на отложенной тестовой выборке
y_test_pred = best_model.predict_proba(X_test)[:, 1]
test_roc_auc = roc_auc_score(y_test, y_test_pred)

print(f"ROC-AUC лучшей модели на тестовой выборке: {test_roc_auc:.4f}")
```

В качестве основной метрики было принято решение выбрать ROC-AUC по следующим причинам:

Независимость от порога принятия решения
– ROC-AUC оценивает способность модели правильно ранжировать «риски» (вероятности класса 1) на всём диапазоне порогов, а не только при условном пороге 0.5.

Устойчивость к дисбалансу классов
– В вашей задаче один класс целевого признака встречается реже другого «снизилась» 500 vs «не снизилась» 800.
– ROC-AUC стабильнее влияет при неравном количестве примеров в классах, в отличие от простой точности (accuracy), которая может быть завышена при сильном преобладании одного класса.

Использование вероятностных предсказаний
– Метрики на жёстких метках (accuracy, F1) игнорируют информацию о том, насколько модель уверена в своём решении.

Стандарт в задачах ранжирования и кредитного скоринга
– ROC-AUC признанным образом демонстрирует, насколько хорошо алгоритм отличает объекты двух классов.
– В бизнес-проектах, где важно выявлять «высокорискованных» клиентов (отток, снижение активности), именно ранжирование по риску — ключевая задача.

Простота интерпретации и сравнения
– Значение ROC-AUC лежит в диапазоне [0.5; 1].
– Лёгко сравнивать разные модели и настройки по единой, нормированной шкале без привязки к конкретному порогу.

Вывод: ROC-AUC позволяет надёжно оценить способность моделей ранжировать клиентов по вероятности снижения активности, остаётся инвариантной к порогу и нечувствительна к дисбалансу, что делает её оптимальным выбором для проекта.

```python
# Разделим признаки
cat_features = ['сервис', 'информирование', 'популярная_категория']

drop_cols = ['активность', 'id']
num_features = (
    df_all
    .columns
    .difference(drop_cols)        # убираем целевой и id
    .difference(cat_features)     # убираем cat-признаки
    .tolist()
)

# Масштабирование числовых признаков
scaler = StandardScaler()
X_train_num = scaler.fit_transform(X_train[num_features])
X_test_num = scaler.transform(X_test[num_features])

# Кодирование категориальных признаков
encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
X_train_cat = encoder.fit_transform(X_train[cat_features])
X_test_cat = encoder.transform(X_test[cat_features])

# Объединяем числовые и категориальные
X_train_ready = np.hstack([X_train_num, X_train_cat])
X_test_ready = np.hstack([X_test_num, X_test_cat])

# Обучаем модель логистической регрессии
model = LogisticRegression(max_iter=1000)
model.fit(X_train_ready, y_train)

ohe_feature_names = encoder.get_feature_names_out(cat_features).tolist()

# Убираем "популярная_категория_" — оставляем только «Мелкая бытовая техника и электроника» и т.п.
clean_ohe_names = [
    name.replace('популярная_категория_', '') 
    for name in ohe_feature_names
]

# Оборачиваем длинные имена по 20 символов
wrapped_ohe_names = [
    '\n'.join(textwrap.wrap(name, width=20))
    for name in clean_ohe_names
]

# Финальный список feature_names: сначала числовые, потом обёрнутые категориальные
feature_names = num_features + wrapped_ohe_names

# Создаем SHAP Explainer для линейной модели
explainer = shap.LinearExplainer(model, X_train_ready, feature_names=feature_names)

# Вычисляем SHAP-значения для тестовой выборки
shap_values = explainer(X_test_ready)

# Увеличиваем дефолтный размер фигуры для всех следующих графиков
plt.rcParams['figure.figsize'] = (10, 8)

# Строим барплот важности через summary_plot
shap.summary_plot(
    shap_values,
    X_test_ready,
    feature_names=feature_names,
    plot_type="bar",
    max_display=17
)
```

##### как можно использовать эти наблюдения при моделировании и принятии бизнес-решений?

Применение в моделировании:

- Упрощение модели: можно отбросить самые низкозначимые фичи (длительность, некоторые категории, выручку текущего месяца) без существенной потери качества — ускорит обучение и снизит переобучение.

- Фокус на сильных: при дальнейшей настройке моделей оставить в первую очередь фичи по просмотрам страниц, времени на сайте и корзине, а остальные агрегировать или убрать.

Бизнес-рекомендации:

- При выявлении групп риска: клиенты с низким числом просмотренных страниц/категорий и малым временем на сайте — в первую очередь включать в рассылку персональных предложений, поскольку именно у них модель фиксирует повышенный риск снижения активности.

- Оптимизация коммуникаций: усилить маркет-активность (увеличить число касаний) именно для тех клиентов, которые уже просматривают мало, чтобы вернуть их интерес.

- Работа с «висящими» корзинами: дополнительное напоминание или скидка для тех, кто часто оставляет товары неоплаченными.

Итог: модель опирается главным образом на поведенческие метрики взаимодействия с сайтом, а не на абсолютные траты или сервисные категории.

### Сегментация покупателей

#### Выполните сегментацию покупателей. Используйте результаты моделирования и данные о прибыльности покупателей.

Сегментируем покупателей следующим образом:

Разделим их на 4 категории в зависимости от показателя целевого признака (покупательская активность) и прибыльности.

```python
# Предсказанные вероятности снижения для всех клиентов
# Берём лучший конвейер из RandomizedSearchCV
best_pipe = randomized_search.best_estimator_

# Формируем полную матрицу признаков для df_all
X_full = df_all.drop(['активность','id'], axis=1)
X_full_ready = best_pipe.named_steps['preprocessor'].transform(X_full)

# Вычисляем вероятность класса “Снизилась” (это второй класс)
probs = best_pipe.named_steps['model'].predict_proba(X_full_ready)[:, 1]

# Данные о прибыльности
profit_df = money.rename(columns={'прибыль':'profit'})

# Собираем в один DF
seg = pd.DataFrame({
    'id': df_all['id'].values,
    'prob_decrease': probs
}).merge(profit_df, on='id')

# Находим медианы по двум осям
prob_med   = seg['prob_decrease'].median()
profit_med = seg['profit'].median()

# Отмечаем сегменты: high/low risk × high/low value
seg['risk_group']  = np.where(seg['prob_decrease'] >= prob_med, 'high_risk', 'low_risk')
seg['value_group'] = np.where(seg['profit']       >= profit_med, 'high_value','low_value')
seg['segment']     = seg['risk_group'] + '_' + seg['value_group']

# Смотрим, сколько в каждом сегменте и средние значения
result = (
    seg
    .groupby('segment')
    .agg(
        count          = ('id', 'size'),
        mean_prob      = ('prob_decrease', 'mean'),
        mean_profit    = ('profit',       'mean')
    )
    .sort_values('mean_prob', ascending=False)
)
print(result)
```

Мы получили четыре сегмента:

- high_risk_high_value

- high_risk_low_value

- low_risk_high_value

- low_risk_low_value

с числом клиентов и средними показатели внутри каждого сегмента.

Эти сегменты дадут основу для дальнейшего анализа профилей клиентов и стратегий персонализации:

например, high_risk_high_value — ключевая «VIP-группа», требующая немедленных удерживающих акций,

low_risk_low_value — скорее всего, минимум вложений и т. д.

Таким образом, сегментация даёт чёткое представление о том, кому и в каком порядке нужно направлять ресурсы для максимизации возврата инвестиций.
